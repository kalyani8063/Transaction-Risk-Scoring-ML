# Transaction Risk Analysis

This notebook performs simple exploratory data analysis on the sample transaction dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Load the dataset from the data folder.
data = pd.read_csv("../data/transactions.csv")

# Display the first few rows.
data.head()
```

```python
# View basic information about the dataset.
data.info()
```

```python
# View summary statistics for the numeric columns.
data.describe()
```

```python
# Plot how transaction amounts are distributed.
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x="amount", bins=10, kde=True)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.show()
```

```python
# Compare the number of normal and fraudulent transactions.
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x="fraud")
plt.title("Fraud vs Normal Transactions")
plt.xlabel("Fraud Label")
plt.ylabel("Number of Transactions")
plt.show()
```
